```python
mdata2 = mdata2.fillna(0)
```

Checking for null values

```python
mdata2.isnull().sum(axis=0)
```

```python
mdata2.drop('Row Labels',axis=1,inplace=True)
```

Checking for unique values, our target is to get only two unique values in each column.

```python
print(mdata2.nunique())
```

Replacing all the values to 0 and 1

```python
print(mdata2['Namkeen'].value_counts())
print(mdata2['Other Dals'].value_counts())

mdata2 = mdata2.replace(3.0, 1.0)
mdata2 = mdata2.replace(2.0, 1.0)

print(mdata2['Namkeen'].value_counts())
print(mdata2['Other Dals'].value_counts())
```

Converting all the values to boolean values

```python
mdata2 = mdata2.replace(1.0,True)
mdata2 = mdata2.replace(0.0, False)
```

```python
mdata2
```

### Model Training

```python
apriori(mdata2, min_support=0.30)
```

The numbers written in the itemset column in the table represent the products (0-51). Item 0 refers to Almonds, item number 1 refers to Poha, item number 51 refers to Whole Spices.

```python
print("Number of Rules:", len(apriori(mdata2, min_support=0.30)))
```

Now, using the use_colnames=True parameter within the apriori algorithm, we switch from items(products) numbers to item(product) names.

```python
frequent_itemsets = apriori(mdata2, min_support=0.30, use_colnames=True)
frequent_itemsets
```

In the table above, it is seen that single, double and triple itemsets are formed. After we set the min_support value (0.30) and create rules sets, we create the Association Rules table according to the metric we are interested in (confidence, lift, conviction and etc.). Here we chose Confidence as the metric and its value 0.3 (30%).

```python
rules1 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.30)
```

```python
print("Number of Rules : ", len(rules1))
```

```python
rules1 = rules1.sort_values(['confidence'], ascending=False)
rules1
```

Observation: If we examine the line with ID information of 37;

    -The probability of seeing Urad Dal and Sugar items together (support) is 34% (0.34),
    -94% (0.944444) of the people (confidence) who bought the Urad Dal item probably also bought the Sugar item,
    -The sales (lift) of the Sugar item in the shopping carts containing the Urad Dal item increased by 1.88 times,
    -How much higher (leverage) 0.16 is when Urad Dal and Sugar items are purchased together than if they are purchased independently of each other,
    -We can say that Urad Dal and Sugar items are related to each other (conviction) with a value of 9.00.
Now let's add the number of items in the antecedents and consequents parts and see the results:

```python
rules1["antecedent_len"] = rules1["antecedents"].apply(lambda x: len(x))
rules1["consequents_len"] = rules1["consequents"].apply(lambda x: len(x))
rules1
```

We can do what we did for the confidence metric above for other metrics. For the lift metric as an example:

```python
rules2 = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules2 = rules2.sort_values(['lift'], ascending=False)
rules2
```

Filtering for Generated Rule Sets

Filter 1: Let's see the records with an Antecident item length of greater than 1 and a Confidence value greater than 0.50 and a Lift value greater than 1

```python
rules1[(rules1['antecedent_len'] > 1) &
       (rules1['confidence'] >= 0.50) &
       (rules1['lift'] > 1) ].sort_values(['confidence'], ascending=False)
```

Observation: If we examine the line with ID information of 38;

    -The probability of seeing Root Vegetables, Beans and Bananas items together (support) is 30% (0.30),
    -88% (0.882353) of the people (confidence) who bought the Root Vegetables & Beans item probably also bought the Banana item,
    -The sales (lift) of the Banana item in the shopping carts containing the Root Vegetables & Beans item increased by 1.22 times,


Filter 2: Similarly, the records with the Antecedents item name Sugar, sorted by Confidence metric [Z-A]:

```python
rules1[rules1['antecedents'] == {'Sugar'}].sort_values(['confidence'], ascending=False)
```

Observation: On examining all the above records; 
    
    We can say that out of 60%(0.60) of the people (confidence) who bought the Sugar item probably also bought either Banana, Other Dals, Beans, Urad Dall or Root Vegetables.
